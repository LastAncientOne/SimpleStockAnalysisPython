# Simple Finance News Scrape

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://finance.google.com/finance/market_news?ei=IbdpWqiTEYWsugSGh4-4Ag"
html = requests.get(url)
soup = BeautifulSoup(html.content, "html.parser")
```

```python
all_tables = soup.find_all('table')
all_tables
```

```python
import newspaper

yahoo_finance_news = newspaper.build('https://finance.yahoo.com/')
```

```python
for article in yahoo_finance_news.articles:
    print(article.url)
```

```python
for category in yahoo_finance_news.category_urls():
    print(category)
```
