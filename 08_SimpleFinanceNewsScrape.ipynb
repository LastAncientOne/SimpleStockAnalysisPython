```python
import newspaper

yahoo_finance_news = newspaper.build('https://finance.yahoo.com/')
```

```python
for article in yahoo_finance_news.articles:
    print(article.url)
```

```python
for category in yahoo_finance_news.category_urls():
    print(category)
```
